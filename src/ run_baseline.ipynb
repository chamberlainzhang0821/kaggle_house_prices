```python
# 在 Notebook 里的单元格写：
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from utils import load_and_clean_data

# 1. 获取清洗后的数据
X_train, y_train, X_test, test_ids = load_and_clean_data()

# 2. 初始化最简单的多元线性回归模型
model = LinearRegression()

# 3. 量化思维：做本地交叉验证 (K-Fold CV)，看看模型水平，不要盲目去线上提交
# 评价指标是明文规定的 RMSE（对数均方根误差）
cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
rmse_scores = np.sqrt(-cv_scores)
print(f"本地 5 折 CV 的平均 RMSE 得分: {rmse_scores.mean():.4f}")

# 4. 用全量训练集训练模型
model.fit(X_train, y_train)

# 5. 预测并逆转对数转换
log_preds = model.predict(X_test)
final_preds = np.expm1(log_preds) # 逆转 log1p

# 6. 生成提交文件
submission = pd.DataFrame({
    'Id': test_ids,
    'SalePrice': final_preds
})
submission.to_csv('../submissions/baseline_lr.csv', index=False)
print("提交文件已生成在 submissions/ 目录下！")
```
